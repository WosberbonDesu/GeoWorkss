```python
def extract_lat_lng(addres_or_postalcode,data_type = "json"):
    # data_type = "json"
    endpoint = f"https://maps.googleapis.com/maps/api/geocode/{data_type}"
    params = {"address": addres_or_postalcode, "key": api_key}
    url_params = urlencode(params)
    url = f"{endpoint}?{url_params}"
    #print(url_params)
    #print()
    r = requests.get(url)
    if r.status_code not in range(200,299):
        return {}
    
    #print(url)
    return r.json()
    
    
```

```python
def color_select(row):
    if row["lived"] == "yes":
        return "red"
    elif row["lived"] == "no" and  row["visited"] == "yes":
        return "green"
    return "blue"
```

```python
cities["color"] = cities.apply(color_select,axis = 1)
cities.head(3)
```

```python
for _, city in cities.iterrows():
    folium.Marker(
    location = [city["latitude"],city["longitude"]],
        popup = city["name"],
        tooltip = city["name"],
        icon = folium.Icon(color = city["color"],prefix = "fa",icon = "circle")
    ).add_to(map)
map
```
